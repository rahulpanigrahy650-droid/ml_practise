```python
# custom indexing
s2 = pd.Series([23,24,25,26], index = ["adam","eve","charlie","bob"])
print(s2["eve"])  #24
print(s2["bob"])  #26

# vectorized operation
s1 = pd.Series([1,2,3])
s2 = pd.Series([4,5,6])

print(s1+s2)

# mutable value but immutable size

s = pd.Series([1,2,3,4,5])
s[0] = 100

print(s)
changed_s = s.drop(1)
print(changed_s)
print(s)
```

```python
# DtaFrame methods

data = {
    "name" : ["arav",'isha',"rohan","sneha","vikram"],
    "age"  : [25,30,35,40,45],
    "city" : ["delhi","mumbai","banglore","kolkata","chennai"]
}

df = pd.DataFrame(data)

print(df.head())
print(df.tail())
print(df.sample())
print(df.info())
print(df.describe())
print(df.nunique)

print(df.dtypes)
print(df.shape)
print(df.columns)
```

```python
# Indexing and data section

# selecting columns
df["country"]  # returning a series
df[["city" , "aqi"]]  # returning a DataFrame

df.loc[0]  #1st row by label
 
df.loc[0:3]  # row0 to row3 both inclusive in loc


df.loc[0,"country"] # 0 th row  & specific column

# in loc the ending index is inclusive 
# done through labels


```

```python
# using iloc

df.iloc[0]  # 1st row
df.iloc[4:7] 

# in iloc  it is done through index position
```

```python
# Fitering & querying data

df[df["age"] > 30]
df[(df["age"] > 30) & (df["salary"]  >5000)]

```

```python
# df.query()

df.query("age >30 and salry < 70000")

my_country = "in"
df.query("country == @my_country") 
# query return a copy not view
# use backticks for columns
# ude @ for python variable

```

```python
# Data cleaning

# handling missing values

df.isnull()   # true for null values

df.isnull().sum()  # counts missing vals per column

df.dropna()  # drops rows with missing values
df.dropna(axis  = 1)


df.fillna(0)  # fills nan with 0 
df["age"]  = df["age"].fillna(df["age"].mean())

df.ffill()  # forward fill
df.bfill()  # backward fill
```

```python
# Handle duplicated values

df.duplicated()  # find duplicates

df.drop_duplicates()  # Remove duplicates rows

df.duplicated()
df.duplicated("country")
df.duplicated(["country" , "gender"])
```
