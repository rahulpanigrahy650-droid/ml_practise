```python
# fancy indexing
# accesing element using integer array list insted using (:)

arr = np.array([1,2,3,4,5])
idx = [0,1,4]
print(arr[idx]) # print nums at given indices

# bollean masking (or indexing)
print(arr[arr>2])
print(arr[arr %2 == 0])
```

```python
# slicing
arr = np.array([1,2,3,4,5,6,7])

print(arr[2:6])
```

```python
# sliced list is a copy

py_list = [1,2,3,4,5]

cpy_list = py_list[1:4]
cpy_list[1] = 333

print(cpy_list)
print(py_list) # remain same

#sliced array is a view

np_arr  =np.array([1,2,3,4,5])

view_arr = np_arr[1:4] # [2,3,4]
view_arr[1] = 333

print(view_arr)
print(np_arr) # changes

# creating copy for array

copy_arr = np_arr[1:4].copy() #[2,3,4]
copy_arr[2] = 444
print(copy_arr)  
print(np_arr) #[1,2,3,4,5]  # remain same


```

```python
# data type

arr = np.array([1,2,3,4,5])
arr2 = np.array([1.0,2.0,3.0])
arr3  = np.array(["hello","world","prime","ai/ml"])

print(arr.dtype) #int 64
print(arr2.dtype) # float64
print(arr3.dtype) #u


# complex numbers

arr1 =  np.array([2+3j])
arr2 = np.array([5+8j])

print(arr1,arr1.dtype)
print(arr1+arr2)
print(arr2 - arr1)

# objects

arr = np.array(["hello",{1,2,3},{3.14}])
print(arr,arr.dtype)
```

```python
 # standard vector normalization
arr = np.array([[1,2],[3,4]])

mean = np.mean(arr)
std_dev = np.std(arr)
normalized_arr = (arr - mean) /std_dev
print(normalized_arr)
```

```python
# column wise normaliztion
arr = np.array([[1,2],[3,4],[5,6]])
mean = np.mean(arr)

std_dev = np.std(arr)
normalized_arr = (arr - mean) / std_dev

print(normalized_arr)
```

```python
# Useful mathmetical function

# Aggregation Function
arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.argmin(arr))
print(np.max(arr))
print(np.argmax(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
```

```python
#power function

arr = np.array([1,2,3,4,5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python

#Log and Exponential function

print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))
print(np.exp(arr))

```
